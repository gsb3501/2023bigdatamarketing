```python
#포스 데이터

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/data2.csv")
df
```

```python
# 평균 단가 낮은 제품 10개 조회
low_price10 = df.groupby('item_name').mean().sort_values(by='item_price').head(10)
low_price10
```

```python
# 평균 단가 높은 제품 10개 조회
low_price10 = df.groupby('item_name').mean().sort_values(by='item_price', ascending = False).head(10)
low_price10
```

```python
# 가장 인기 많은 메뉴 10개 조회
top10 = df.groupby('item_name').sum().sort_values(by='quantity', ascending = False).head(10)
top10
```

```python
# 가장 인기 없는 메뉴 10개 조회
top10 = df.groupby('item_name').sum().sort_values(by='quantity').head(10)
top10
```
