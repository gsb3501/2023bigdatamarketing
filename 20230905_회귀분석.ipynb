```python
# Y 선정 (내가 궁금한 것 : 1회 평균매출액을 높이려면 어떻게 해야할까?)
Y = df['1회 평균매출액']
Y
```

```python
# X 선정 (1회 평균매출액에 영향을 주는 것들이 어떤 것들이 있을까?)
num = ['총매출액', '할인권 사용 횟수', '총 할인 금액', '평균 구매주기']
cg = ['구매금액대', '고객등급', '거주지역' ]
```

```python
X = df[num+cg]
X
```

```python
# Y의 분포 확인 : 왼쪽으로 치우쳤구만~
sns.distplot(Y)
```

```python
# Y를 정규분포로 변환
Y = np.log1p(Y)
sns.distplot(Y)
```

```python
# 표준화 / 원핫인코딩 : 실수 -> 단위를 날려서 분포를 맞춰줌 / 집단변수로 인식시키기 위함

ct = ColumnTransformer([('scaling', StandardScaler(), num),('onehot', OneHotEncoder(sparse = False), cg)])
ct
```

```python
ct.fit(X)
```

```python
X = ct.transform(X)
```

```python
X = pd.DataFrame(X)
X
```

```python
X.columns = ['총매출액', '할인권 사용 횟수', '총 할인 금액', '평균 구매주기', '구매금액대0', '구매금액대1', '구매금액대2',
             '고객등급1', '고객등급2', '거주지역1', '거주지역2', '거주지역3', '거주지역4', '거주지역5', '거주지역6', '거주지역7']
X
```

```python
data = pd.concat([X, Y], axis = 1)
data
```

```python
# ols를 돌려보면 X변수 중 다중공선성이 있는 지 확인할 수 있음
reg = ols(formula = 'Y~X', data = data).fit()
```

```python
# 데이터 스플릿
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)

print(X_train.head())
print('----------------------------------')
print(X_test.head())
print('----------------------------------')
print(Y_train.head())
print('----------------------------------')
print(Y_test.head())
```

# 선형 회귀

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
Y_pred_lr = lr.predict(X_test)

print(Y_pred_lr)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
X.columns
```

```python
coefficient = pd.DataFrame(lr.coef_, X.columns, columns = ['coefficient'])
coefficient
```

```python
coef = pd.DataFrame(lr.coef_)
coef
```

```python
X_col = pd.DataFrame(X.columns)
X_col
```

```python
coef2 = pd.concat([X_col, coef], axis =1)
coef2
```

```python
print("R2 : ", lr.score(X_train, Y_train)) # f(X)로부터 나온 R2
# R2 : 평균보다 얼마나 더 잘 설명하느냐 (0에 가까우면 평균과 똑같다. 1에 가까우면 평균보다 데이터를 더 잘 설명한다는 뜻)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(Y_pred_lr, Y_test)
print("MSE : ", mse ) #MSE는 작을 때 좋음 (오차가 적다는 뜻)
```

```python
# rmse : root mean_squared_error

from math import sqrt

rmse = sqrt(mse)
print("rmse : ", rmse)
```

```python
cor_lr = pd.DataFrame(Y_pred_lr, Y_test)
cor_lr.corr()
```

# 릿지, 라쏘, 엘라스틱

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
```

```python
rg = Ridge(alpha = 0.5)
ls = Lasso(alpha = 0.1)
en = ElasticNet(alpha = 0.5)
```

```python
rg.fit(X_train, Y_train)
ls.fit(X_train, Y_train)
en.fit(X_train, Y_train)
```

```python
Y_pred_rg = rg.predict(X_test)
Y_pred_ls = ls.predict(X_test)
Y_pred_en = en.predict(X_test)

print(Y_pred_rg)
print("------------------------------------------------")
print(Y_pred_ls)
print("------------------------------------------------")
print(Y_pred_en)
```

```python
print("Ridge R2 : ", rg.score(X_train, Y_train))
print("Lasso R2: "  , ls.score(X_train, Y_train))
print("Elastic R2 : ", en.score(X_train, Y_train))
```

```python
rgmse = mean_squared_error(Y_pred_rg, Y_test)
lsmse = mean_squared_error(Y_pred_ls, Y_test)
enmse = mean_squared_error(Y_pred_en, Y_test)


print("Ridge MSE : ", rgmse)
print("Lasso MSE : ", lsmse)
print("Elastic MSE : ", enmse)
```

```python
# rmse : root mean_squared_error

rmse = sqrt(mse)

print("Ridge MSE : ", sqrt(rgmse))
print("Lasso MSE : ", sqrt(lsmse))
print("Elastic MSE : ", sqrt(enmse))
```
