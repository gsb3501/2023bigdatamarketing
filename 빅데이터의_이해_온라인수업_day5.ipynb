# 분산분석 (ANOVA)
집단이 3개 이상일 때 실행

```python
df
```

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import statsmodels.stats.multicomp as mc
```

```python
aov = ols(formula = 'sepal_length~C(Target)', data = df).fit()
anova_lm(aov)
```

# 사후검정

```python
comp = mc.MultiComparison(df["sepal_length"], df["Target"])
tukeyhsd = comp.tukeyhsd(alpha = 0.05)
tukeyhsd.summary()
```

```python
tukeyhsd.plot_simultaneous()
```

```python
# 상관관계 확인
df.corr()
```

```python
from sklearn.linear_model import LinearRegression # 선형회귀
from sklearn.model_selection import train_test_split # random sampling
from sklearn.metrics import mean_squared_error # mse: 에러는 작을수록 좋음
```

```python
df.info()
```

```python
cols = list(df.columns)
cols
```

```python
x_cols  = cols[1:4]
y_cols = cols[0]
```

```python
x_cols
```

```python
y_cols
```

```python
iris_df = df[cols[:4]]

iris_df
# 지금 sepal_length 가 Y(종속변수), sepal_width~ petal_width 는 X(독립변수)
```

```python
# train set, test set  나누기

iris_train, iris_test = train_test_split(iris_df, test_size = 0.3)
```

```python
iris_train.info()
```

```python
iris_test.info()
```

```python
model = LinearRegression()
model.fit(X = iris_train[x_cols], y = iris_train[y_cols])
```

```python
model.coef_ #상관계수
```

```python
model.intercept_ # 상수항
```

```python
iris_train.info()
```

```python
# test set 의 x값만 가지고 y값을 모델로 만들기

y_pred = model.predict(X = iris_test[x_cols])
y_pred
```

```python
# y 실제값

iris_test["sepal_length"]
```

```python
# 잔차 구하기 (에러값)
mse = mean_squared_error(y_true = iris_test["sepal_length"], y_pred = y_pred)
print("mse =", mse)
```

```python
correlation = pd.DataFrame({"Y" : iris_test["sepal_length"], "Y_pred": y_pred})
correlation
```

```python
correlation.corr()

# 현재 모델의 예측력이 상당히 높음
```
