```python
import math
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from scipy.stats import boxcox
df_join['총매출액_scale'] = preprocessing.scale(df_join["총매출액"]) #scale : 표준화 (단위를 없애주는 조치) ex) z표준화
df_join.head()
```

```python
scaler  = MinMaxScaler()  #표준화
x_scale = df_join[['1회 평균매출액', '총 할인 금액']] #내가 scale 하고 싶은 것 1회 평균매출액, 총 할인금액만 꺼냈음
x_scale_minmax = scaler.fit_transform(x_scale) #x_scale 을 minmaxscaler로 표준화해줘
x_scale_minmax = pd.DataFrame(x_scale_minmax) # minmaxscaler 를 적용하면 array로 반환되어서 이것을 dataframe 형태로 바꿨음
x_scale_minmax.head()
```

```python
x_scale_minmax.describe()
```

```python
df_join_minmax = pd.concat([df_join, x_scale_minmax], axis = 1)
#엑셀 형태의 두 개 (df_join, x_scale_minmax)를 열 방향 (axis=1)으로 붙여줘.
df_join_minmax.head()
```

```python
df_join_minmax.columns
```

```python
df_join_minmax.columns = [      '고객ID',       '이탈여부',       '총매출액',      '구매금액대',       '방문빈도',
         '1회 평균매출액',  '할인권 사용 횟수',    '총 할인 금액',       '고객등급',       '구매유형',
          '클레임접수여부',    '구매카테고리수',       '거주지역',         '성별',     '고객 나이대',
             '거래기간',     '할인민감여부',    'Recency',  'Frequency',   'Monetary',
          '평균 구매주기', '총매출액_scale',  '1회평균액_minmax', '총 할인금액_minmax']
df_join_minmax.head()
```
